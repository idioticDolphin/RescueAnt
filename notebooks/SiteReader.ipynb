```python
import json
from enum import Enum
```

```python
class Station:
    def __init__(self, name:str, address, animals:list, mail:str, phone:str, url:str, found_at:str):
        """
        Create a new Station Object

        :param name: name of the station
        :param address: address of the station
        :param animals: type of animals taken by the station
        :param mail: e-mail address of the station
        :param phone: phone number of the station
        """
        self.name = name
        self.address = address
        self.animals = animals
        self.mail = mail
        self.phone = phone
        self.url = url
        self.found_at = found_at

    def save_json(self, filename:str, save_type="w") -> None:
        """
        Serializes and saves the Station Object to a JSON file
        :param filename: path of the save file
        :param save_type: specify whether to overwrite ('w') or append('a') existing files
        """
        if not (save_type == "w" or save_type == "a"):
            raise ValueError("save_type must be 'w' or 'a'")

        with open(filename, save_type) as outfile:
            json.dump(self.to_dict(), outfile, indent=4)

    def to_dict(self) -> dict:
        """
        Serializes the Station Object to a dict
        :return: dict representation of the Station Object
        """
        return {
            "name": self.name,
            "address": self.address,
            "animals": self.animals,
            "mail": self.mail,
            "phone": self.phone,
            "url": self.url,
            "found_at": self.found_at
        }

    @classmethod
    def from_dict(cls, data:dict) -> 'Station':
        """
        Creates a Station Object from a dictionary
        :param data: serialized dict representation of a Station Object
        :return: deserialized Station Object
        """
        return Station(data['name'], data['address'], data['animals'], data['mail'], data['phone'], data['url'], data['found_at'])

    @classmethod
    def read_json(cls, filename:str) -> Station:
        """
        Reads a Station Object from a JSON file
        :param filename: path of the JSON file
        :return: deserialized Station Object
        """
        with open(filename, "r") as json_file:
            data = json_file.read()
        return cls.from_dict(json.loads(data))
```
